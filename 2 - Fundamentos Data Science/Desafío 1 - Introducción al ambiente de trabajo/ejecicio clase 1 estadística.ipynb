```python
import pandas as pd 
import numpy as np 

df_flights=pd.read_csv('flights.csv')
print(df_flights.head(5))
print(df_flights[-5:])

print(df_flights['year'].describe())
print(df_flights['year'].value_counts())
print(df_flights['month'].value_counts())

df_primeras15=df_flights[:15]
df_ultimas15=df_flights[-15:]
print(df_primeras15)
print(df_ultimas15)

mediatotal=np.mean(df_flights['passengers'])
medianatotal=(np.median(df_flights['passengers']))
desviacióntotal=(np.std(df_flights['passengers']))
sumatotal=df_flights['passengers'].sum()

media15=np.mean(df_primeras15['passengers'])
mediana15=(np.median(df_primeras15['passengers']))
desviación15=(np.std(df_primeras15['passengers']))
suma15=df_primeras15['passengers'].sum()

media15final=np.mean(df_ultimas15['passengers'])
mediana15final=(np.median(df_ultimas15['passengers']))
desviación15final=(np.std(df_ultimas15['passengers']))
suma15final=df_ultimas15['passengers'].sum()
cv_total=desviacióntotal/mediatotal
cv_primeras=desviación15/media15
cv_final=desviación15final/mediana15final


print("La media total la muestra es de {} pasajeros". format(mediatotal))
print("La mediana total la muestra es de {} pasajeros". format(medianatotal))
print("La desviación estándar de la muestra es de {} pasajeros". format(desviacióntotal))
print("La cantiadad total de pasajeros de la muestra es de {} \n". format(sumatotal))

print("La media de las primeras 15 observaciones de la muestra es de {} pasajeros". format(media15))
print("La mediana de las 15 primeras observaciones de la la muestra es de {} pasajeros". format(mediana15))
print("La desviación estándar de las primeras 15 obsevaciones es de {} pasajeros". format(desviación15))
print("La cantiadad total de pasajeros de las primeras 15 observaciones de la muestra son {} \n". format(suma15))

print("La media de las últimas 15 observaciones de la muestra es de {} pasajeros". format(media15final))
print("La mediana de las 15 últimas observaciones de la la muestra es de {} pasajeros". format(mediana15final))
print("La desviación estándar de las últimas 15 obsevaciones es de {} pasajeros". format(desviación15final))
print("La cantiadad total de pasajeros de las últimas 15 observaciones de la muestra son {} \n". format(suma15final))

print("El coeficiente de Variabilidad de la muestra total es de {}". format(cv_total))
print("El coeficiente de Variabilidad de las primeras 15 observaciones de la muestra es de {}". format(cv_primeras))
print("El coeficiente de Variabilidad de las últimas 15 observaciones de la muestra es de {}". format(cv_final))

# Las últimas 15 observaciones de la muestra presentan una media de 459,2 pasajeros mensuales cifra significativamente mayor 
# a la media de la muestra total que presenta 280,29 pasajeros mensuales  y a la media de  las primeras 15 observaciones,
# las que presentan la menor media con solo 126,8 pasajeros. Por otro lado, las primeras 15 observaciones presentan un menor 
# nivel de dispersión, con un coeficiente de variabilidad de 10% respecto de su media, en contraste con la variabilidad de la
#muestra total que presenta un coeficiente del 42,65% sobre su media.
```

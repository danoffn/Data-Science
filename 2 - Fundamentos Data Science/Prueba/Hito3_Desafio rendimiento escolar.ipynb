## Evaluación de factores latentes

En primer lugar vamos a inspeccionar una batería de preguntas asociadas a aspectos ambientales del alumno (de famrel a health) para ver si éstas se pueden abstraer en categorías latentes.

Para ello aplicaremos un análisis factorial que nos permita estudiar dichos factores latentes.

__Variables a considerar__:

$famrel$ : Calidad de las relaciones familiares. (numérica: de 1 - muy malas a 5 - excelentes)

$freetime$ : Tiempo libre fuera del colegio (numérica: de 1 - muy poco a 5 - mucho)

$goout$ : Salidas con amigos (numérica: de 1 - muy pocas a 5 - muchas)

$Dalc$ : Consumo de alcohol en día de semana (numérica: de 1 - muy bajo a 5 - muy alto)

$Walc$ : Consumo de alcohol en fines de semana (numérica: de 1 - muy bajo a 5 - muy alto)

$health$ : Estado de salud actual (numérica: from 1 - muy malo to 5 - muy bueno)


```python
#Seleccionamos las variables a incluir en el análisis factorial

df_fact = df_selec[['famrel', 'freetime', 'goout', 'Dalc', 'Walc', 'health']].copy()
print("La base de datos tiene {} variables y {} registros.".format(df_fact.shape[1],df_fact.shape[0]))
```

```python
#evaluemos la media de cada variable para analizar una posible agrupación de las variables

variable = []
mean = []

for colname, serie in df_fact.iteritems():
    variable.append(colname)
    mean.append(serie.mean())
    
df_1 = pd.DataFrame({'var':variable, 'mean': mean}).sort_values(by=['mean'])
plt.plot(df_1['mean'], df_1['var'],'o');
```

Observamos en un extremo, con la media más baja las variables que hacen referencia al consumo de alcohol indicando "Muy bajo consumo". Por otro lado, hacia el extremo superior se ubica la variable Calidad de Relaciones Familiares indicando que la media se aproxima a la evaluación de excelentes. 

En un punto intermedio, se ubican de forma cercana la media de las variables "Salida con Amigos" y "Tiempo Libre".

Ahora, antes de modelar los factores latentes evaluaremos la ideonidad de nuestra matriz de datos para extraer factores. Para ello aplicaremos las pruebas de esferacidad de Barlett y de muestreo de Kaiser-Meyer-Olkin


```python
print("Chi-cuadrado= {}".format(factor.calculate_bartlett_sphericity(df_fact)[0]))
print("P-Value= {}".format(factor.calculate_bartlett_sphericity(df_fact)[1]))
```

Obtenemos de acuerdo a la prueba de Esfericidad de Barlet un __p-value < 0.05__, de esta forma la prueba de esferacidad sugiere que la matriz de correlaciones no es igual a la matriz de identidad, por lo que se puede seguir con el análisis.

Continuamos con la prueba de muestreo Kaiser-Meyer-Olkin (KMO).

```python
factor.calculate_kmo(df_fact)
```

Observamos que el valor de KMO esta por debajo del punto de corte aceptable (0.7). Por lo que se plantea una nueva matriz de datos excluyendo la variable con menor puntaje de KMO y repetimos las pruebas de validación.

# Regresiones lineales modelos saturados

Procedemos a aplicar tres regresiones lineales, considerando (en modelos separados) las variable G1, G2 y G3 como variables dependientes y la bateria de variables completa como las variables independientes.

Previamente, debemos considerar que las variables categóricas necesitan un tratamiento previo al análisis, en el cual se convierten en variables dummy o cualitativas tomando valores binarios por cada columna. Es decir, por cada valor que toma la variable categórica se creará una columna del valor y tomara valor 1 si la instancia presenta ese valor de la variable categórica y 0 en caso contrario. Además, para simplificar el conjunto de datos se elimina las variables convertidas en dummy, puesto que no aportan nada al modelo.

```python
df_selec_dum = pd.get_dummies(df_selec,drop_first=True)

print("La base de datos dummy tiene {} variables y {} registros.".format(df_selec_dum.shape[1],df_selec_dum.shape[0]))
```

```python
# Generamos la lista de variables independientes a considerar en las regresiones

cols = df_selec_dum.columns.tolist()
cols.remove("G1")
cols.remove("G2")
cols.remove("G3")
string_vars ='+'.join(cols)

print(string_vars)
```

### Modelo 1: para G1, notas durante el primer semestre

```python
model_g1 = smf.ols('G1 ~ school_bin+sex_bin+age+address_bin+famsize_bin+Pstatus_bin+Medu+Fedu+traveltime+studytime+failures+schoolsup_bin+famsup_bin+paid_bin+activities_bin+nursery_bin+higher_bin+internet_bin+romantic_bin+famrel+freetime+goout+Dalc+Walc+health+absences+Mjob_health+Mjob_other+Mjob_services+Mjob_teacher+Fjob_health+Fjob_other+Fjob_services+Fjob_teacher+reason_home+reason_other+reason_reputation+guardian_mother+guardian_other', data=df_selec_dum)
model_g1 = model_g1.fit()
model_g1.summary()
```

Dado el valor de  __R-Cuadrado = 0.331__ se determina que para nuestro modelo ejecutado, la variabilidad de las notas durante el primer semestre se explica en un __33.1%__ por las variables incluidas en el modelo. Además, tenemos el valor de __Prob(F-statistic) < 0.05__, por lo que tenemos evidencia en contra de la hipótesis nula y hay más evidencia a favor de la significancia estadística general del modelo.

__Depuración del modelo:__ en miras de mejorar el modelo de regresión, analizaremos las correlaciones del conjunto de variables independientes para evitar la la posible redundancia entre variables y procedemos a realizar una nueva regresión.

```python
plt.figure(figsize=(25,25))
sns.heatmap(df_selec.corr(), cmap='Blues', annot=True);
```

En primer lugar se observa la alta correlación entre las variables G1, G2 y G3, lo cual era esperable dado que las tres miden el promedio de notas en distintos momentos del semestre.

En segundo lugar, las variables con mayor correlación son "Dalc" vs "Walc", las cuales hacen referencia al consumo del alcohol tanto en día de semana como el fin de semana. Esta correlación es relativamente alta, sin embargo no genera una redundancia importante entre las variables.

Dado lo anterior, para depurar el modelo solo generaremos una nueva correlación con aquellas variables con __p-value < 0.05__.

```python
model_g1_dep = smf.ols('G1 ~ studytime + failures + schoolsup_bin + famsup_bin + freetime + goout + health' , data=df_selec_dum)
model_g1_dep = model_g1_dep.fit()
model_g1_dep.summary()
```

Para el modelo depurado, obtenemos un valor de  __R-Cuadrado = 0.221__ lo que indica que este modelo reduce la variabilidad de las notas durante el primer semestre se explica, siendo ahora un __22.1%__ por las variables incluidas en este segundo modelo. Además, tenemos el valor de __Prob(F-statistic) < 0.05__, por lo que tenemos evidencia en contra de la hipótesis nula y hay más evidencia a favor de la significancia estadística general del modelo.

Veamos que ocurre para el modelo con las notas del segundo semestre.
